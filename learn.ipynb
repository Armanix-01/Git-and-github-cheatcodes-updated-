```python
#GIT DIFF
"""git giff - is  an informative command that shows the differences'ArithmeticError
between two commits. It is used to compare the changes made in 
one commit with changes made in another commit.  Git consider the
changes version of same file as two different files. Then it gives
names to these two files and shows the diiffernce between them."""
## Comparing staging area with repository.
"""git diff --staged

What is staging area?
 Think of it like a shopping cart:
 You edit files ‚Üí they are in the working directory.
 You choose which changes to commit ‚Üí you put them in the 
 staging area using git add.
 Finally, you commit the staged changes ‚Üí they become part of 
 Git history.
  
 Here we compare our staging area(what are the things you added
 using add and what were the things before added)
"""
## Comparing between two branches
"""git diff branch_name1..branch_name2"""#(preferred)
"""git diff branch1 branch2"""
## Comparing specific commits
"""git diff <commit-hash-one>..<commit-hash-two>"""#(preferred)
"""git diff <commit-hash-one> <commit-hash-two>"""
```

```python
#GIT STASH
"""stash is a way to save your changes in a temp location. It is
useful when you want to make changes to a file but dont want to commit
them yet. you can then come back to the file later and commit the
changes. what it allows us to made some change and save it temp
and switch branch, normally when we switch branch it will not allow
us to do as conflict will arise but with git stash we can do it"""
## Syntax
"""git stash"""
## naming the stash
"""git stash save "work in progress on X feature" """
## List of stash
"""git stash list"""
## Apply the stash(it means the work you had done is in stash and
#you want to make it back to the branch).
"""git stash apply"""
## Apply the specific stash
""" git stash apply stash@{0} """
## Applying and then dropping the stash
""" git stash pop"""
## Drop the stash
"""git stash drop"""
## Applying stash to a particular branch
"""git stash apply stash@{0} <branch-name>"""
## Clearing the stash
"""git stash clear"""
```

```python
"""When you run:

git pull --rebase origin main


Git will update your local branch with changes from the remote
 branch using rebasing instead of a merge. Let‚Äôs break it down clearly.

üîÑ What happens step by step

Git temporarily removes your local commits that are not yet on 
the remote.
Example:

Remote (origin/main): A ‚Äî B ‚Äî C
Your local (main):   A ‚Äî B ‚Äî C ‚Äî D ‚Äî E


Your commits D and E are local only.

Git updates your branch to match the remote

Your branch now temporarily: A ‚Äî B ‚Äî C


Git re-applies your local commits on top of the updated remote

After rebase: A ‚Äî B ‚Äî C ‚Äî D' ‚Äî E'


Notice D' and E' ‚Äî these are your commits ‚Äúrebased‚Äù on top of the
 remote commits."""
```

```python
##Remote branch important info
"""1Ô∏è‚É£ Remote branches

When you clone a repo, Git creates copies of the branches from 
GitHub (or another remote) on your local machine.
These copies are called remote-tracking branches.
Example: origin/main ‚Üí this is your local reference to the 
main branch on GitHub.
Think of it as a mirror of the remote branch.

2Ô∏è‚É£ Local branches

When you start working, you usually have a local branch: main
This is the branch you actually work on ‚Äî you make commits here.

3Ô∏è‚É£ Tracking a remote branch

You can tell your local branch to ‚Äúfollow‚Äù a remote branch:
git branch -u origin/main
Now your local main branch tracks origin/main.

This means:

Action	   | Result
git push   | Sends local commits to origin/main automatically
git pull   | Pulls changes from origin/main automatically

You don‚Äôt have to type origin main every time.

üîπ Simple analogy

Remote branch (origin/main) = mirror on GitHub

Local branch (main) = your copy to work on

Tracking (git branch -u) = connects your copy to the mirror, so push/pull happens automatically."""
## after git branch -u origin/main just mention git pull and git push
```
